# INTRODUÇÃO AO PANDAS 

```python
import pandas as pd
```

# METODOS DE IMPORTAÇÃO DATAFRAME
- feito localmente
- importado de excel

```python
# data frame [ LOCAL ]
venda = {'Data': ['15/02/2024', '16/02/2024'],
         'Valor': [1200, 800],
         'Produto': ['SSD 120GB', 'SSD 256GB'],
         'Quantidade': [10, 3],
         }
vendas_df = pd.DataFrame(venda)
display(vendas_df)
```

```python
# data frame [ EXCEL ] ( ARQUIVO PRECISA ESTAR NO MESMO DIRETORIO! )
vendas_df = pd.read_excel("Vendas.xlsx")
display(vendas_df)
```

# RESUMO DE VISUALIZAÇÕES
- head()            [ VISUALIZAÇÃO DAS PRIMEIRAS "X" LINHAS (df.head(X)). ]
- shape             [ VISUALIZAÇÃO DE COLUNAS E LINHAS DA TABELA. ]
- describe()        [ DESCRIÇÃO DE ESTATISTICAS. ]

```python
# HEAD ↓
display(vendas_df.head())

# SHAPE ↓
display(vendas_df.shape)

# DESCRIBE ↓
display(vendas_df.describe())
```

# EXIBIÇÃO/SELEÇÃO DE COLUNAS
- 1 coluna
- 2 ou mais colunas

```python
# 1 COLUNA
produtos = vendas_df['Produto']
display(produtos)

# 2 OU MAIS COLUNAS ( 2 COLCHETES )
produtos = vendas_df[['Produto', 'Valor Final']]
display(produtos)
```

# METODO DE LOCALIZAÇÃO
- pegar 1 linha
- pegar linhas de acordo com condição
- pegar linhas e colunas especificas
- pegar 1 valor especifico

```python
# exemplo.loc[X] [ X = NÚMERO DO INDICE, PRIMEIRA LINHA DO DATAFRAME ]

# pegar uma linha
display(vendas_df.loc[0])

# pegar intervalo de linhas
display(vendas_df.loc[0:5])
```

```python
# pegar linhas de acordo com condições [ APENAS LOJAS NORTE SHOPPING ]
vendas_norteshopping = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping)
```

```python
# pegar varias linhas e colunas usando loc
# vendas_norteshopping = vendas_df.loc[linhas, colunas]

vendas_norteshopping = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ["ID Loja", "Produto", "Quantidade"]]
display(vendas_norteshopping)
```

```python
# pegar valor especifico
# exemplo.loc[linha, coluna]

print(vendas_df.loc[1, 'Produto'])
```

# ADICIONAR COLUNAS
- adicionar coluna com informação já existente na tabela
- adicionar coluna com valores manualmente

```python
# a partir de uma coluna que existe [ criar coluna de comissão * 0.05 ]
# EX: tabela_df['NOME_NOVA_COLUNA'] = tabela_df['COLUNA_EXISTENTE'] * 0.05

vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
# criar coluna nova [ coluna imposto com um valor declarado manualmente ]
# EX: tablea_df.loc[:, "NOME_NOVA_COLUNA"] = VALOR_DECLARADO
vendas_df.loc[:, "Imposto"] = 0
display(vendas_df)
```

# ADICIONAR NOVAS LINHAS | CONCATENAR TABELAS
- concatenando linhas de uma tabela a outra [ junção de tabelas ]
- novas linhas

```python
# IMPORTANDO NOVA TABELA COM MAIS INFORMAÇÕES ↓
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")

# METODO CONCAT PRA JUNTAR AS TABELAS  [ SE RODAR MAIS DE UMA VEZ VAI ADICIONAR NOVAMENTE]
# EX: tabela_antiga = pd.concat([tabela_antiga, TABELA_NOVA])
vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

# EXCLUIR LINHAS
- metodo drop [ axis = 1 (COLUNA) | axis = 0 (LINHA) ]

```python
# excluir coluna de imposto
vendas_df = vendas_df.drop("Imposto", axis=1)
display(vendas_df)
```

# CALCULAR INDICADORES
- value counts
- groupby

```python
# value counts
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)

#NOME DA LOJA | TRANSAÇÕES REALIZADAS PELA LOJA
```

```python
# !!!!!!! GROUPBY NÃO SUPORTA DATAS !!!!!!!
# vendas_df = vendas_df.drop('Data', axis=1)

#groupby [ TABELA INTEIRA ]
#faturamento_produto = vendas_df.groupby('Produto').sum()
#display(faturamento_produto)

# groupby [ COLUNAS SELECIONADAS ]
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

# MESCLAR DATAFRAMES
- metodo merge

```python
# importando nova tabela
gerentes_df = pd.read_excel('Gerentes.xlsx')
# display(gerentes_df)

# adicionando tabela de gerentes na coluna principal
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
