```python
# STEP 1/2/3
import pandas as pd

drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks
```

```python
# STEP 4.1
continent4 = drinks.groupby(by=['continent']).agg(Brejas=('beer_servings', 'mean'))
continent4 = continent4.sort_values(by=['Brejas'],ascending=False)
continent4
```

```python
# STEP 4.2
continent_media = drinks.groupby(by=['continent']).agg(Brejas=('beer_servings', 'mean'))
continent_media = continent_media.sort_values(by=['Brejas'], ascending=False)
continent_media = continent_media.drop(columns=['Brejas']).head(1)
continent_media
```

```python
# STEP 5
vinhos = drinks.groupby(by=['continent']).agg(Soma=('wine_servings', 'sum'), Media=('wine_servings', 'mean'), Maior=('wine_servings', 'max'), Menor=('wine_servings', 'min'))
vinhos
```

```python
#STEP 6
alcool = drinks.groupby(by=['continent']).total_litres_of_pure_alcohol.mean()
print(alcool)
```

```python
# STEP 8
spirit = drinks.groupby(by=['continent']).agg(Media=('spirit_servings', 'mean'), Maior=('spirit_servings', 'max'), Menor=('spirit_servings', 'min') )
spirit
```
