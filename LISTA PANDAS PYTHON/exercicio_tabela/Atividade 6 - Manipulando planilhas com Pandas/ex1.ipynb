```python
# INICIO EXERCICIO LOGICA DE PROGRAMAÇÃO

import pandas as pd
import numpy as np

caminho = (r'C:\Users\joao_A_moura\Desktop\python\senaipy\exercicio_tabela\vendas.xlsx')
df = pd.read_excel(caminho)
```

```python
# EXERCICIO 1
df['ValorUnitario'] = round(df['ValorLiquido'] / df['Qtde'],2)
df_valorprodutos = df[['NomeProduto', 'ValorUnitario']].drop_duplicates()
df_valorprodutos = df_valorprodutos.sort_values(by='NomeProduto')
df_valorprodutos = df_valorprodutos.reset_index(drop=True)
df_valorprodutos.insert(0, 'IdProduto', range(1, len(df_valorprodutos)+1))
df_valorprodutos
```

```python
# EXERCICIO 2 
df_ordem_cliente = df[['NomeCliente']].drop_duplicates()
df_ordem_cliente = df_ordem_cliente.reset_index(drop=True)
df_ordem_cliente = df_ordem_cliente.sort_values(by='NomeCliente')
df_ordem_cliente.insert(0, 'IdCliente', range(1, len(df_ordem_cliente)+1))
df_ordem_cliente
```

```python
#EXERCICIO 3
tabela_nota = df[['NotaFiscal', 'NomeCliente']].drop_duplicates()
tabela_nota.sort_values(by='NotaFiscal')
tabela_nota.reset_index(drop=True)
tabela_nota.insert(0, 'IDNF', range(1, len(tabela_nota) +1))
tabela_nota = tabela_nota.merge(df_ordem_cliente, on=['NomeCliente'])
tabela_nota = tabela_nota.drop(columns=['NomeCliente'])
tabela_nota
```
